```python
from pathlib import Path
import ollama
import pandas as pd

method = 'old'

basepath = Path('processed_txts_' + method)
files_in_basepath = basepath.iterdir()

d = {}
for item in files_in_basepath:
    file = open('processed_txts_' + method + '/' + item.name, "r")
    content = file.read()
    emb = ollama.embeddings(model='gemma2:27b', prompt=content)
    d[item.name] = emb
    file.close()
    
```

```python
df = pd.DataFrame(d)
df = df.T
```

```python
df.to_parquet('df_embs_'+ method + '_gemma2.parquet')
```
